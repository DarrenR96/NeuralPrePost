```python
from library import * 
import torch 
from torchsummary import summary 
from torchviz import make_dot

config_data = load_toml_file('configs/models/trini_base.toml')

model = TRINIModel(**config_data)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

x_comp = torch.randn(1, 1, 256, 256, device=device)
x_ref = torch.randn(1, 1, 256, 256, device=device)

out = model(x_ref, x_comp)
dot = make_dot(out, params=dict(model.named_parameters()))

# Multiple inputs: pass a list of (C, H, W) shapes, one per model argument
summary(model, [(1, 256, 256), (1, 256, 256)], device=device)
dot.render('simple_net.png')
```
